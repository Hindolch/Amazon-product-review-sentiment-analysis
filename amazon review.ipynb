```python
ps = PorterStemmer()
```

```python
def transform_text(text):
    text = text.lower()
    text = nltk.word_tokenize(text)
    x = []
    for i in text:
        if i.isalnum():
            x.append(i)
    text = x[:]
    x.clear()
    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
            x.append(i)
    text = x[:]
    x.clear()
    for i in text:
        x.append(ps.stem(i))
    return " ".join(x)
```

```python
transform_text('yo whats going on? aint you coming to gym today?')
```

```python
df['transformed'] = df['reviewText'].apply(transform_text)
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
x = cv.fit_transform(df['transformed']).toarray()
y = df['overall'].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, multilabel_confusion_matrix
```

```python
x_train, x_temp, y_train, y_temp = train_test_split(x,y, test_size=0.35, random_state=100)
x_valid, x_test, y_valid, y_test = train_test_split(x_temp, y_temp, test_size=0.35, random_state=100)
```

```python
model = MultinomialNB()
```

```python
model.fit(x_train, y_train)
```

```python
pred = model.predict(x_valid)
print(accuracy_score(y_valid, pred))
print(precision_score(pred, y_valid, average='weighted'))
```

```python
import pickle
pickle.dump(cv, open('countVec.pkl', 'wb'))
pickle.dump(model, open('model.pkl', 'wb'))
```
